```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
data = pd.read_csv("titanic.csv")
```

```python
sns.countplot(x='Survived', data=data)
plt.title("Overall Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(data['Age'], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=data)
plt.title("Age vs Survival")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()
```

```python
sns.histplot(data['Fare'], bins=40, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=data)
plt.title("Fare vs Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.title("Survival by Embarkation Port")
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
survival_rate = data.groupby(['Sex','Pclass'])['Survived'].mean().reset_index()

sns.barplot(x='Pclass', y='Survived', hue='Sex', data=survival_rate)
plt.title("Survival Probability by Class and Gender")
plt.ylabel("Survival Probability")
plt.show()
```
